2) Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. name, rating, year of release) and make data frame.

```python
imdb1 = requests.get('https://www.imdb.com/search/title/?groups=top_100&sort=user_rating,desc')
imdb2 = requests.get('https://www.imdb.com/search/title/?groups=top_100&sort=user_rating,desc&start=51&ref_=adv_nxt')
```

```python
imdb1_soup = BeautifulSoup(imdb1.content)
imdb2_soup = BeautifulSoup(imdb2.content)
```

```python
name = []
for i in imdb1_soup.find_all('h3' ,class_="lister-item-header" ):
    name.append(i.text)
for i in imdb2_soup.find_all('h3' ,class_="lister-item-header" ):
    name.append(i.text)
name_n = []
for element in name:
    name_n.append(element.strip())
name_n
```

```python
rating = []
for i in imdb1_soup.find_all('div' ,class_="inline-block ratings-imdb-rating" ):
    rating.append(i.text)
for i in imdb2_soup.find_all('div' ,class_="inline-block ratings-imdb-rating" ):
    rating.append(i.text)
    
rating_n = []
for element in rating:
    rating_n.append(element.strip())
rating_n
```

```python
import pandas as pd
```

```python
dataframe = pd.DataFrame({'Movie Name':name_n,'rating':rating_n})
```

7) Write a python program to scrape details of all the posts from coreyms.com. Scrape the heading, date, content
and the code for the video from the link for the youtube video from the post.

```python
tutorial = requests.get("https://coreyms.com/")
tutorial_soup = BeautifulSoup(tutorial.content)
```

```python
header = []
for i in tutorial_soup.find_all('a',class_="entry-title-link"):
    header.append(i.text)

date = []
for i in tutorial_soup.find_all('time',class_="entry-time"):
    date.append(i.text)

content = []
for i in tutorial_soup.find_all('div',class_="entry-content"):
    content.append(i.text)

content_clean = []
for element in content:
    content_clean.append(element.strip())

ytlink = []
for i in tutorial_soup.find_all('figure',class_="wp-block-embed-youtube wp-block-embed is-type-video is-provider-youtube wp-embed-aspect-16-9 wp-has-aspect-ratio"):
    ytlink.append(i.link)
ytlink
```

```python
df_core = pd.DataFrame({"Heading":header,"Date":date,"content":content_clean})
df_core
```
